```python
from music21 import *
from music21.midi.realtime import StreamPlayer
import numpy as np
import music21
from matplotlib import pyplot as plt
from ipywidgets import interact, fixed
import numpy as np
import random
```

```python
def create_emotion_music(emotion):
    # Get the musical elements for this emotion
    emotion_music = emotion_to_music[emotion]

    # Create a new music21 stream
    s = stream.Stream()

    # Set the tempo
    s.append(tempo.MetronomeMark(number=emotion_music['tempo']))

    # Add the notes
    for note_name in emotion_music['notes']:
        n = note.Note(note_name)
        dyn = dynamics.Dynamic(emotion_music['dynamics'])
        n.expressions.append(dyn)
        s.append(n)

    # Return the stream
    return s
```
